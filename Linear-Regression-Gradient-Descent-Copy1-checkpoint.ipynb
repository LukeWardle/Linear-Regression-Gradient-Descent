The aim of this project is to build a simple linear regression model to predict output values based on input features. I will be using gradient descent to to optimize performance.

The data used in this project will be synthetic, the parameters will be updated using gradient descent and I will be using visuals to demonstrate and assess model fit.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# For reproducibility
np.random.seed(42)
```

I have imported numpy and matplotlib. Numpy will be used for numeric operations and handling arrays, while matplotlib will be used to visualize the data

```python
# Define the number of data points
m = 100  # number of examples

# Generate random X values between 0 and 2 (shape: 100 x 1)
X = 2 * np.random.rand(m, 1)

# Generate y values based on a linear relationship: y = 4 + 3*X + noise
y = 4 + 3 * X + np.random.randn(m, 1)
```

X is our synthetic data it is a feature matrix with 100 (m) examples. Y is our target feature computed as y = 4 + 3X with additional random noise.

Noise ensures that the model must learn to handle small fluctuations or errors‚Äîthis makes the exercise more realistic and helps prevent overfitting to an overly ‚Äúclean‚Äù dataset.

Each 
ùë¶
y value is calculated by starting with 4, then adding 3 times the 
ùëã
X value, and finally adding a small random number. This random number (noise) makes the data less perfect and more like real-life data."

```python
# Add a column of ones to X for the bias term
X_b = np.c_[np.ones((m, 1)), X]
```

X_b is the augmented feature matrix where the first column (ones) represents the bias term, making it easier to learn the intercept along with the other weights

```python
# Initialize theta randomly (shape: 2 x 1 for bias and slope)
theta = np.random.randn(2, 1)

# Set the hyperparameters for gradient descent
learning_rate = 0.1
n_iterations = 1000

```

theta represents the parameters of our model (including the bias and the weight for X). The learning rate determines the step size in each update, and the number of iterations controls how many updates we perform.

```python
cost_values = [] 
# Gradient descent loop to update theta
for iteration in range(n_iterations):
    # Calculate the predictions
    predictions = X_b.dot(theta)
    
    # Compute the error (difference between predictions and actual values)
    error = predictions - y
    
    # Calculate the cost (Mean Squared Error)
    cost = np.mean(error ** 2)
    cost_values.append(cost)
    
    # Compute the gradient using the transpose of X_b
    gradients = 2/m * X_b.T.dot(error)
    
    # Update theta by moving in the direction opposite to the gradient
    theta = theta - learning_rate * gradients

# Output the final parameters
print("Estimated parameters (theta):", theta)
```

At each iteration, we compute the prediction error and then calculate the gradient, which tells us how to update theta. The formula for updating is theta := theta - learning_rate √ó gradient.

X_b ‚Üí Prediction (X_b ¬∑ theta) ‚Üí Error (Prediction - y) ‚Üí Gradient (X_b·µÄ ¬∑ Error) ‚Üí Update theta

```python
# Create a new figure for plotting
plt.figure(figsize=(8, 6))

# Plot the original data points
plt.scatter(X, y, color='blue', label='Data Points')

# Create a sequence of values for X to plot the regression line
X_new = np.linspace(0, 2, 100).reshape(100, 1)
X_new_b = np.c_[np.ones((100, 1)), X_new]

# Calculate the predicted y values using the learned theta
y_predict = X_new_b.dot(theta)

# Plot the regression line
plt.plot(X_new, y_predict, color='red', linewidth=2, label='Regression Line')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Fit with Gradient Descent')
plt.legend()
plt.show()


plt.figure(figsize=(8, 4))
plt.plot(cost_values, color='green')
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Cost Function Convergence (lr = 0.1)")
plt.show()
```

The blue dots represent the original data, while the red line shows the model's
predictions after training. Visualizing the fit helps us see if the model is 
learning the correct relationship.

The cost function starts at about 30 and rapidly decreases to nearly 0, indicating fast convergence. The nearly vertical drop at the final iteration is likely due to the cost being almost 0 by that point and the plotting scale, meaning the model has effectively learned the data pattern.

```python
# Experiment: Hyperparameter Comparison

# Define different hyperparameter settings
learning_rates = [0.01, 0.1, 0.5]         # example learning rates
iterations_list = [1000, 5000]              # different iteration counts

# Create a grid for subplots: rows for learning rates, columns for iteration counts
fig, axs = plt.subplots(len(learning_rates), len(iterations_list), figsize=(14, 10))

# Loop over each combination of learning rate and iterations
for i, lr in enumerate(learning_rates):
    for j, n_iter in enumerate(iterations_list):
        theta = np.random.randn(2, 1)  # reinitialize theta for each experiment
        cost_values = []

        # Run gradient descent for this combination
        for iteration in range(n_iter):
            predictions = X_b.dot(theta)
            error = predictions - y
            cost = np.mean(error ** 2)
            cost_values.append(cost)
            gradients = 2/m * X_b.T.dot(error)
            theta = theta - lr * gradients
        
        # Plot the cost function convergence
        ax = axs[i, j]
        ax.plot(cost_values, color='green')
        ax.set_title(f"lr = {lr}, iter = {n_iter}")
        ax.set_xlabel("Iterations")
        ax.set_ylabel("Cost (MSE)")
        ax.grid(True)

plt.tight_layout()
plt.show()
```

### Experiment Observations

- **Learning Rate 0.1, 1000 Iterations:** The cost function drops sharply and levels off near 0.
- **Learning Rate 0.01, 1000 Iterations:** The cost decreases more gradually, indicating smaller updates.
- **Learning Rate 0.5, 1000 Iterations:** The model diverges, as seen by an unstable cost curve.
- **Increasing Iterations:** At 5000 iterations, a low learning rate (0.01) eventually shows a smoother decrease in cost, confirming that more iterations can help convergence when using small updates.

These charts help me compare how different learning rates and iteration counts affect convergence.
